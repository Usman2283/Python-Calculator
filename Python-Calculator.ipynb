```python
# Calculator
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

while True:
    print("\nCalculator")
    print("1: Add")
    print("2: Subtract")
    print("3: Multiply")
    print("4: Divide")
    print("5: Exit")
    choice = input("Choose operation: ")
    if choice == "5":
        break
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    if choice == "1":
        print("Result: ", add(num1, num2))
    elif choice == "2":
        print("Result: ", sub(num1, num2))
    elif choice == "3":
        print("Result: ", mul(num1, num2))

    elif choice == "4":
        print("Result: ", div(num1, num2))
```
